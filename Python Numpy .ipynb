```python
import numpy as np

arr1 = np.array([12, 13, 14, 15, 16, 17])
arr2 = np.array([24, 25, 27, 28, 29, 30])

A = np.add(arr1, arr2)
S = np.subtract(arr1, arr2)
M = np.multiply(arr1, arr2)
D = np.divide(arr1, arr2)
P = np.power(arr1, arr2)

print("Addition:",A)
print("subtraction:",S)
print("multiplication:",M)
print("Division:",D)
print("Power",P)
```
